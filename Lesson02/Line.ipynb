```python
df = pd.read_csv('data/dow_jones_index.data')
df.stock.unique()
df.index = pd.to_datetime(df.date)
df[df.stock.isin(['MSFT', 'GE', 'PG'])].groupby('stock')['volume'].plot()
```

```python
_ = df.groupby('stock').volume.plot()
```

```python
fig, ax = plt.subplots()
_ = df.groupby('stock').volume.plot(subplots=True, ax = ax)
```

```python
fig, ax = plt.subplots()
for key, group in df.groupby('stock'):
    group.volume.plot(ax=ax)
```
